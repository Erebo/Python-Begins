```python
x=int(input())
y=int (input())
add=x+y
subtraction =x-y
multiplication = x*y
division = round(x/y,2)
remainder = x%y
print(add,subtraction,multiplication,division,remainder)
```

```python
# The comparison operators always return in boolyan 
# Bool means either 1 or 0 , True or False 
x = int(input())
y = int(input())
greater_than = x>y
greater_than_or_equal = x>=y
less_than = x<y
print(greater_than,greater_than_or_equal,less_than)
```
